```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Paths to your image folders
train_folder = r'C:\Users\User\Documents\My University Heriot Watt\Data Mining and machine learning\Final project\Data sets\images\train'
validation_folder = r'C:\Users\User\Documents\My University Heriot Watt\Data Mining and machine learning\Final project\Data sets\images\validation'

# Function to display images from a folder, including subdirectories
def display_images(folder):
    # Walk through the directory
    for root, dirs, files in os.walk(folder):
        count = 0  # Counter for displayed images

        for file_name in files:
            # Construct full file path
            file_path = os.path.join(root, file_name)

            # Check if the file is an image (you can expand this list as needed)
            if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                try:
                    img = Image.open(file_path)
                    plt.imshow(img)
                    plt.title(file_name)
                    plt.axis('off')  # Hide axes
                    plt.show()

                    count += 1  # Increment count of displayed images

                    # Stop after displaying 5 images
                    if count >= 5:
                        break

                except Exception as e:
                    print(f"Could not open {file_name}: {e}")

        # Stop displaying images after showing 5 images
        if count >= 5:
            break

# Display images from both train and validation folders
print("Training Images:")
display_images(train_folder)

print("Validation Images:")
display_images(validation_folder)
```

```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Paths to your image folders
train_folder = r'C:\Users\User\Documents\My University Heriot Watt\Data Mining and machine learning\Final project\Data sets\images\train'
validation_folder = r'C:\Users\User\Documents\My University Heriot Watt\Data Mining and machine learning\Final project\Data sets\images\validation'

# Function to display images and print pixel descriptions
def display_images_with_pixel_info(folder):
    # Walk through the directory
    for root, dirs, files in os.walk(folder):
        count = 0  # Counter for displayed images

        for file_name in files:
            # Construct full file path
            file_path = os.path.join(root, file_name)

            # Check if the file is an image
            if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                try:
                    img = Image.open(file_path)

                    # Convert image to numpy array for pixel access
                    img_array = np.array(img)

                    # Display the image
                    plt.imshow(img)
                    plt.title(file_name)
                    plt.axis('off')  # Hide axes
                    plt.show()

                    # Print pixel information
                    print(f"Image: {file_name}")
                    print(f"Shape: {img_array.shape}")  # Shape of the image (height, width, channels)
                    print(f"Pixel values (top-left corner):\n{img_array[:5, :5]}")  # Print pixel values for a 5x5 region
                    print("\n" + "="*40 + "\n")

                    count += 1  # Increment count of displayed images

                    # Stop after displaying 5 images
                    if count >= 5:
                        break

                except Exception as e:
                    print(f"Could not open {file_name}: {e}")

        # Stop displaying images after showing 5 images
        if count >= 5:
            break

# Display images and pixel information from both train and validation folders
print("Training Images:")
display_images_with_pixel_info(train_folder)

print("Validation Images:")
display_images_with_pixel_info(validation_folder)
```
