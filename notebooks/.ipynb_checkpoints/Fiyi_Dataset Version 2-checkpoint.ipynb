```python
#Function to load data
def data_loader():
    path_url = 'https://raw.githubusercontent.com/HWhr3000/F21DL_Coursework_grp2/main/data/airline_passenger_satisfaction.csv'
    return pd.read_csv(path_url)
    
data = data_loader()
data.head()
```

```python
# Calculate the correlation matrix focusing on Satisfaction, Loyalty, Seat Comfort, and Online Boarding
important_columns = ['Satisfaction', 'Loyalty', 'Seat Comfort', 'Online Boarding']
corr_matrix = data[important_columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Loyalty and Related Features')
plt.show()

# Scatter plot for Seat Comfort vs Loyalty
data.plot(kind='scatter', x='Seat Comfort', y='Loyalty', alpha=0.5)
plt.title('Seat Comfort vs Loyalty')
plt.xlabel('Seat Comfort')
plt.ylabel('Loyalty')
plt.show()

# Scatter plot for Online Boarding vs Loyalty
data.plot(kind='scatter', x='Online Boarding', y='Loyalty', alpha=0.5)
plt.title('Online Boarding vs Loyalty')
plt.xlabel('Online Boarding')
plt.ylabel('Loyalty')
plt.show()

# Boxplot for Loyalty across Online Boarding levels
sns.boxplot(x='Online Boarding', y='Loyalty', data=data)
plt.title('Loyalty Distribution across Online Boarding Levels')
plt.show()
```

```python
heatmap_data = data.groupby(['Seat Comfort', 'Loyalty']).size().unstack()
sns.heatmap(heatmap_data, annot=True, cmap="Blues")
plt.title('Heatmap of Seat Comfort vs Loyalty')
plt.show()

sns.stripplot(x='Seat Comfort', y='Loyalty', data=data, jitter=True)
plt.title('Seat Comfort vs Loyalty with Jitter')
plt.show()
```

<h1> Experimenting with Attribute Combinations </h1>

```python
# New Feature 1: Service Quality Score (sum of key service aspects)
data['service_quality_score'] = (
    data['Check-in Service'] + 
    data['Online Boarding'] + 
    data['On-board Service'] + 
    data['Seat Comfort'] + 
    data['In-flight Entertainment'] + 
    data['Cleanliness']
)

# New Feature 2: Delay Impact Score (combines departure and arrival delays)
data['delay_impact_score'] = (
    data['Departure Delay'] + data['Arrival Delay'].fillna(0)  # Handling missing arrival delays
)

# New Feature 3: Comfort vs. Duration Ratio (Seat Comfort relative to Flight Distance)
data['comfort_vs_duration'] = data['Seat Comfort'] / data['Flight Distance']

# New Feature 4: Total In-flight Experience (sum of in-flight services)
data['total_inflight_experience'] = (
    data['In-flight Wifi Service'] + 
    data['In-flight Entertainment'] + 
    data['In-flight Service'] + 
    data['Food and Drink']
)

# Checking the first few rows of the newly created features
print(data[['service_quality_score', 'delay_impact_score', 'comfort_vs_duration', 'total_inflight_experience']].head())
```

```python
# Recalculate the correlation matrix after adding the new features
corr_matrix = data.corr()

# Sort and print the correlations of all attributes with 'Loyalty'
loyalty_correlations = corr_matrix["Loyalty"].sort_values(ascending=False)
print(loyalty_correlations)
```
