```python
import sys
print(sys.executable)

import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Set parameters
img_size = (48, 48)  # Resize all images to 48x48
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,  # Add zooming
    brightness_range=[0.8, 1.2],  # Adjust brightness
    shear_range=0.2,  # Shear transformations
    horizontal_flip=True
)

# Only rescale for validation
val_datagen = ImageDataGenerator(rescale=1.0/255.0)

# Load images from separate directories
data_train_dir = '../data/train'
data_val_dir = '../data/validation'

train_generator = train_datagen.flow_from_directory(
    data_train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    data_val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)
```

```python
# Load images from train directory and flatten them for PCA
data = []
labels = []

# Iterate through each emotion category folder in the train directory
for folder_name in os.listdir(data_train_dir):
    folder_path = os.path.join(data_train_dir, folder_name)
    if os.path.isdir(folder_path):  # Ensure it's a directory
        for img_name in os.listdir(folder_path):
            img_path = os.path.join(folder_path, img_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                img = cv2.resize(img, img_size)
                img_rgb = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)  # Convert to RGB
                data.append(img_rgb)  # Add RGB image to data list
                labels.append(folder_name)  # Store label as the folder name

data = np.array(data)
labels = np.array(labels)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=50)
data_pca = pca.fit_transform(data.reshape(len(data), -1))  # Flatten for PCA

# Apply KMeans clustering
kmeans = KMeans(n_clusters=7)  # Assume 7 clusters for 7 emotions
kmeans.fit(data_pca)

# Visualize clusters
plt.figure(figsize=(10, 8))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("Emotion Clusters")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
# Load pre-trained VGG16 model with custom input shape, excluding the top layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(48, 48, 3))

# Add custom classification layers on top of the VGG16 base model
x = base_model.output
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(7, activation='softmax')(x)  # 7 output classes for emotions

model = Model(inputs=base_model.input, outputs=output)

# Freeze all layers in the VGG16 base model to retain pre-trained weights
for layer in base_model.layers:
    layer.trainable = False

# Compile the model with a suitable optimizer and loss function for multi-class classification
model.compile(optimizer=Adam(learning_rate=0.00001), loss='categorical_crossentropy', metrics=['accuracy'])

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)

# Fit the model
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,  # Adjust as needed
    steps_per_epoch=train_generator.samples // batch_size,
    validation_steps=val_generator.samples // batch_size,
    callbacks=[early_stopping]
)
```

```python
# Evaluate the model's performance on the validation set
val_loss, val_accuracy = model.evaluate(val_generator)
print(f"Validation Loss: {val_loss:.4f}")
print(f"Validation Accuracy: {val_accuracy:.4f}")

# Plot training and validation accuracy over epochs
plt.figure(figsize=(10, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy', color='b')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='r')
plt.title("Model Accuracy During Training")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
